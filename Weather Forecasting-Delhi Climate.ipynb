```python
# Importing python libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
# data link: <https://www.kaggle.com/datasets/sumanthvrao/daily-climate-time-series-data>
path = r"C:\Users\Administrator\Documents\PythonProjects Jupyter\delhi climate dataset\DailyDelhiClimateTrain.csv"
data = pd.read_csv(path)
```

```python
# Importing the dataset needed
def delhi_weather_data(df):
    print(df.tail(5))
    print(" ")
    print(df.describe())
    print(" ")
    print(df.info())
```

```python
output = delhi_weather_data(data)
output
```

```python
# Mean temperature in Delhi over the years
def mean_temp_vs_year(df):
    mean_temp = df["meantemp"]
    date = df["date"]
    plt.style.use('Solarize_Light2')
    plt.figure(figsize=(15,10))
    plt.plot(date, mean_temp, color='b')
    plt.xlabel('Date')
    plt.ylabel('Mean Temperature')
    plt.title('Mean Temperature in Delhi Over the Years')
    plt.xticks(['2013-01-01', '2013-06-01', '2014-01-01', '2014-06-01', '2015-01-01', '2015-06-01',
                '2016-01-01', '2016-06-01', '2017-01-01'],
               ['Jan 2013', 'Jun 2013', 'Jan 2014', 'Jun 2014', 'Jan 2015',
                'Jun 2015', 'Jan 2016', 'Jun 2016', 'Jan 2017'])
    plt.show()
```

```python
output = mean_temp_vs_year(data)
output
```

```python
# Humidity in Delhi over the years
def humidity_vs_year(df):
    humidity = df["humidity"]
    date = df["date"]
    plt.style.use('Solarize_Light2')
    plt.figure(figsize=(15,10))
    plt.plot(date, humidity)
    plt.xlabel('Date')
    plt.ylabel('Humidity')
    plt.title('Humidity in Delhi Over the Years')
    plt.xticks(['2013-01-01', '2013-06-01', '2014-01-01', '2014-06-01', '2015-01-01', '2015-06-01',
                '2016-01-01', '2016-06-01', '2017-01-01'],
               ['Jan 2013', 'Jun 2013', 'Jan 2014', 'Jun 2014', 'Jan 2015',
                'Jun 2015', 'Jan 2016', 'Jun 2016', 'Jan 2017'])
    plt.show()
```

```python
output = humidity_vs_year(data)
output
```

```python
# Wind Speed in Delhi Over the Years
def wind_speed_vs_year(df):
    wind_speed = df["wind_speed"]
    date = df["date"]
    plt.style.use('Solarize_Light2')
    plt.figure(figsize=(15, 10))
    plt.plot(date, wind_speed, color='g')
    plt.xlabel('Date')
    plt.ylabel('Wind Speed')
    plt.title('Wind Speed in Delhi Over the Years')
    plt.xticks(['2013-01-01', '2013-06-01', '2014-01-01', '2014-06-01', '2015-01-01', '2015-06-01',
                '2016-01-01', '2016-06-01', '2017-01-01'],
               ['Jan 2013', 'Jun 2013', 'Jan 2014', 'Jun 2014', 'Jan 2015',
                'Jun 2015', 'Jan 2016', 'Jun 2016', 'Jan 2017'])
    plt.show()
```

```python
output = wind_speed_vs_year(data)
output
```

```python
# Relationship Between Temperature and Humidity
def temperature_vs_humidity(df):
    humidity = df["humidity"]
    mean_temp = df["meantemp"]
    plt.style.use('Solarize_Light2')
    plt.figure(figsize=(15,10))
    plt.scatter(humidity, mean_temp, c=mean_temp, cmap='summer')
    cbar = plt.colorbar()
    cbar.set_label('Temperature Level')
    plt.xlabel('Humidity')
    plt.ylabel('Mean Temperature')
    plt.title('Relationship Between Temperature and Humidity')
    plt.show()
```

```python
output = temperature_vs_humidity(data)
output
```

```python
# Temperature Change in Delhi Over the Years
def temp_change_in_years(df):
    df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    plt.style.use('Solarize_Light2')
    plt.figure(figsize=(15, 10))
    plt.title("Temperature Change in Delhi Over the Years")
    sb.lineplot(data=df, x='month', y='meantemp', hue='year')
    plt.show()
```

```python
output = temp_change_in_years(data)
output
```
